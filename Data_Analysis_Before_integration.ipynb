# Data Analysis of our datasets information before integration

```python
import pandas as pd
```

```python
df_neigh = pd.read_csv(r"E:\Bahare (important)\uni\Master\Data Managenement\Datasets\austin_zip_to_neighborhood_scraped.csv")
df_pop = pd.read_csv(r"E:\Bahare (important)\uni\Master\Data Managenement\Datasets\austin_population_by_zip_scraped.csv")
df_street_zip = pd.read_csv(r"E:\Bahare (important)\uni\Master\Data Managenement\Datasets\austin_street_to_zip_mapping.csv")
df_court = pd.read_csv(r"E:\Bahare (important)\uni\Master\Data Managenement\Datasets\Municipal_Court_Caseload_Information_FY_2023.csv")
```

```python
df_neigh["ZIP_Code"] = df_neigh["ZIP_Code"].astype(str).str.strip()
df_pop["ZIP_Code"] = df_pop["ZIP_Code"].astype(str).str.strip()
df_street_zip["POSTCODE"] = df_street_zip["POSTCODE"].astype(str).str.strip()

df_street_zip["STREET"] = df_street_zip["STREET"].str.upper().str.strip()
df_court["Offense Street Name"] = df_court["Offense Street Name"].str.upper().str.strip()
```

## Neighborhood → ZIP Dataset 

```python
#How many distinct zipcodes in this dataset we have? 
total_zip_neigh = df_neigh["ZIP_Code"].nunique()
print("Distinct ZIPs in neighborhood:", total_zip_neigh)
```

```python
# Do we have duplicated zipcodes in this dataset? 
dup_zip_neigh = df_neigh[df_neigh.duplicated("ZIP_Code", keep=False)]
dup_zip_neigh["ZIP_Code"].unique()
```

## Population → ZIP dataset

```python
# Do we have duplicated zipcodes in this dataset? 
dup_zip_pop = df_pop[df_pop.duplicated("ZIP_Code", keep=False)]
dup_zip_pop["ZIP_Code"].unique()
```

```python
# The zipcodes with zero population (unknown)
zero_pop = df_pop[df_pop["Population"] == 0]
zero_pop_zips = zero_pop["ZIP_Code"].unique()

print("ZIPs with population = 0:", len(zero_pop_zips))
zero_pop_zips
```

```python
# valid zipcode with population
valid_pop = df_pop[df_pop["Population"] > 0]
valid_pop_zips = valid_pop["ZIP_Code"].unique()

print("ZIPs with valid population:", len(valid_pop_zips))
valid_pop_zips
```

## Street → ZIP dataset

```python
# How many distinct streets we have in this dataset? 
dup_streets = df_street_zip[df_street_zip.duplicated("STREET", keep=False)]
dup_streets["STREET"].unique()
```

```python
# The number of distinct streets in this dataset
distinct_streets = dup_streets["STREET"].unique()
print("The number of distinct streets :", len(distinct_streets))
```

```python
# How many distinct zipcodes do we have in this dataset?
distinct_zip_street = df_street_zip["POSTCODE"].nunique()
print("Distinct ZIPs in street-to-zip:", distinct_zip_street)
```

## Court dataset

```python
# How many distinct streets we have ? And what are their names ? 
distinct_court_streets = df_court["Offense Street Name"].dropna().unique()

print("Distinct offense streets:", len(distinct_court_streets))
distinct_court_streets[:30]  #showing only the first 30 of them
```

## Common ZipCodes between the three datasets 

```python
zip_neigh = set(df_neigh["ZIP_Code"])
zip_pop_valid = set(valid_pop["ZIP_Code"])
zip_street = set(df_street_zip["POSTCODE"])

common_zips = zip_neigh & zip_pop_valid & zip_street

print("Common ZIPs:", len(common_zips))
common_zips
```

```python
streets_court = set(df_court["Offense Street Name"].dropna())
streets_mapping = set(df_street_zip["STREET"])

common_streets = streets_court & streets_mapping

print("Common streets:", len(common_streets))
list(common_streets)[:30]
```

## Intersection Analysis: Validating Common Streets within Target Zip Codes

```python
# Intersection between streets.zip and the common_streets
street_zip_common = df_street_zip[
    df_street_zip["STREET"].isin(common_streets)
]
```

```python
# # Intersection between street_zip_common and the common_zips
street_zip_common_valid = street_zip_common[
    street_zip_common["POSTCODE"].isin(common_zips)
]
streets_with_valid_zip = street_zip_common_valid["STREET"].unique()

print("Number of common streets with valid ZIP:", len(streets_with_valid_zip))
streets_with_valid_zip[:30]  
```

```python
# 
percentage = (
    len(streets_with_valid_zip) / len(common_streets) * 100
)

print(f"{percentage:.2f}% of common streets map to ZIPs with valid population and neighborhood")
```
